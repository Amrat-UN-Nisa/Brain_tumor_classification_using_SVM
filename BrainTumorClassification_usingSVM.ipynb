#Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
# create a LabelEncoder object
encoder = LabelEncoder()
# fit the encoder to the data and transform the data
df['y'] = encoder.fit_transform(df['y'])
# print the encoded data
print(df['y'])
```

```python
df['y'].unique()
```

```python
X = df.drop('y',axis=1)
Y= df['y']
```

# Split data into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
#can ensure that the same sequence of random numbers is generated each time you run the code, which helps you to obtain the same results every time.
```

```python
from sklearn.svm import SVC
# Create an SVM classifier with a linear kernel
clf = SVC(kernel='linear')
```

```python
# Train the classifier using the training data
clf.fit(X_train, y_train)
```

```python
y_train_predict = clf.predict(X_train)
```

```python
y_train_predict.shape
```

```python
y_test_predict = clf.predict(X_test)
```

```python
y_test_predict.shape
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_test_predict)
```

```python
df.columns
```

```python
from sklearn.metrics import r2_score 
accuracy = r2_score(y_test, y_test_predict)
```

```python
# Print the accuracy
print("Accuracy:", accuracy*100)
```
